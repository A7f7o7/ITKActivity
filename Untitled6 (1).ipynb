```python
import itk
import numpy as np
import matplotlib.pyplot as plt

import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
```

```python
img = itk.imread(r"C:\Users\14404\Downloads\coronal.nii")
```

```python
print(img.shape)
```

```python
metadata = itk.dict_from_image(img)
```

```python
metadata
```

```python
img_np2 = np.asarray(img)

fig = plt.figure(figsize=(10,20))

ax = fig.add_subplot(1,3,1)
imgplot = plt.imshow(img_np2[:,:,500])
ax = plt.gca()
#ax.set_aspect((int(metadata['size'][1])/int(metadata['size'][2]))* int(metadata['spacing'][0]))
#ax.set_aspect(10)
ax.set_aspect((metadata['spacing'][2])/(metadata['spacing'][0]))
ax.set_title('Coronal view')


ax = fig.add_subplot(1, 3, 2)
imgplot = plt.imshow(img_np2[10,:,:])
ax.set_title('Transversal view')


ax = fig.add_subplot(1, 3, 3)
imgplot = plt.imshow(img_np2[:,:,50])
ax = plt.gca()
#ax.set_aspect((int(metadata['size'][1])/int(metadata['size'][2]))*int(metadata['spacing'][2]))
#ax.set_aspect(10)
ax.set_aspect((metadata['spacing'][2])/(metadata['spacing'][1]))
ax.set_title('Saggital view')
```
